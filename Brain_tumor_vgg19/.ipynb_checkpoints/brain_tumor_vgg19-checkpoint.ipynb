```python
import os
from cv2 import imread
import matplotlib.pyplot as plt

import keras
from keras.layers import Flatten, Dense, Dropout
#from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import ImageDataGenerator


from tensorflow.keras.applications import ResNet50, VGG19

import warnings
warnings.filterwarnings("ignore")
```

```python
path = "brain_tumor_dataset"

image_generator = ImageDataGenerator(
    rescale=1./255, 
    horizontal_flip=2, 
    vertical_flip=2, 
    validation_split=0.1)    

training_set = image_generator.flow_from_directory(batch_size=16,
                                                 directory=path,
                                                 shuffle=True,
                                                 target_size=(224,224), 
                                                 subset="training",
                                                 color_mode='rgb',
                                                 class_mode='binary')

testing_set = image_generator.flow_from_directory(batch_size=16,
                                                 directory=path,
                                                 shuffle=True,
                                                 target_size=(224,224), 
                                                 subset="validation",
                                                 color_mode='rgb',
                                                 class_mode='binary')
```

```python
yes_path = path+"/yes/"+os.listdir(path+"/yes/")[4]
tumour = imread(yes_path)
plt.imshow(tumour)
```

```python
no_path = path+"/no/"+os.listdir(path+"/no/")[4]
tumour = imread(yes_path)
plt.imshow(tumour)
```

```python
testing_set.class_indices
```

```python
model = VGG19(
      input_shape = (224,224,3),
      include_top = False,
      weights = 'imagenet'
    )
```

```python
for layers in model.layers:
    layers.trainable = False
```

```python
x = Flatten()(model.output)
x = Dropout(0.4)(x)
x = Dense(1, activation = "sigmoid")(x)

model = keras.Model(model.input, x)
#model.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = "accuracy")
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

model.summary()
```

```python
#hist = model.fit_generator(training_set,validation_data = testing_set, epochs = 60)
hist = model.fit(training_set, validation_data=testing_set, epochs=20)
```

```python
hist = hist.history
```

```python
plt.plot(hist["accuracy"])
plt.plot(hist["val_accuracy"])
plt.title("Accuracy plot")
plt.legend(["train","test"])
plt.xlabel("epoch")
plt.ylabel("accuracy")
```

```python
plt.plot(hist["loss"])
plt.plot(hist["val_loss"])
plt.title("Accuracy loss")
plt.legend(["train","test"])
plt.xlabel("epoch")
plt.ylabel("loss")
```

```python
evaluation = model.evaluate(testing_set)

# Print the accuracy
accuracy = evaluation[1]
print(f'Test Accuracy: {accuracy * 100:.2f}%')
```
